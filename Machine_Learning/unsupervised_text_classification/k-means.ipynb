```python
import numpy as np
import re
```

```python
with open('text.txt','r') as reader:
    text_data = reader.read()
text_data = text_data.lower()
split_data = text_data.replace(r"\s","").split(".")
d = re.sub(r"\.\s+",".",text_data).split(".")
if "" in d:
    d.remove("")
d
```

```python
import nltk
#nltk.download("stopwords")
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()
```

```python
stemmed_dataset = []
for i in range(0, len(d)):
    stemmed_array = d[i].split(" ")
    stemmed = [ps.stem(word) for word in stemmed_array if not word in set(stopwords.words("english"))]
    stemmed = " ".join(stemmed)
    stemmed_dataset.append(stemmed)
```

```python
stemmed_dataset
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
lemmatizer_dataset = []
for i in range(0, len(d)):
    lemmatizer_array = d[i].split(" ")
    lemmatizer = [lemma.lemmatize(word) for word in lemmatizer_array if not word in set(stopwords.words("english"))]
    lemmatizer = " ".join(lemmatizer)
    lemmatizer_dataset.append(lemmatizer)
```

```python
lemmatizer_dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(lemmatizer_dataset)
X
```

```python
from sklearn.cluster import KMeans
wcss =[]
for i in range(0, len(lemmatizer_dataset)):
    kmeans = KMeans(n_clusters=1,init='k-means++',max_iter=300,n_init=10,random_state=0,verbose=True)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
from matplotlib import pyplot as plt
plt.plot(range(0,len(lemmatizer_dataset)),wcss)
plt.title('The elbow method')
plt.xlabel('no. of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
true_k = 10
model = KMeans(n_clusters=true_k,init='k-means++',n_init=1)
model.fit(X)
```

```python
print('Top terms per cluster')
order_centroids = model.cluster_centers_.argsort()[:,::-1]
terms = cv.get_feature_names()
for i in range(true_k):
    print('cluster %d :'%i)
    for ind in order_centroids[i,:10]:
        print(' %s ' %terms[ind]), print
print('\n')
    
```

```python
Y = cv.transform(["returned"])
prediction = model.predict(Y)
print(prediction)
```
